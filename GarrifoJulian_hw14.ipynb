```python
#############################
#                           #
#          3                #
#                           #
#############################

def semicirculo_unitario(angulo):
    r = 1.0 # radio unitario
    x = r*cos(angulo) # Valores x de cada angulo
    y = r*sin(angulo) # Valores y de cada angulo
    return [x,y]

angulo = arange(0,pi,0.05) # Crear angulos desde 0 hasta pi, cada 0.05 radianes
[x,y] = semicirculo_unitario(angulo) # Obtener los valores x,y del cemicirculo

# Graficar el circulo
figure()
plot(x,y,c='k',lw=2,label='dist')
axis('equal') # Aegurarse de que la grafica tenga relacion de aspecto 1:1 para que el semi-circulo se vea como tal

# Crear arreglos de x,y,r con 1000 posiciones para guardar las muestras aleatorias
x = zeros(1000)
y = zeros(1000)
r = zeros(1000)
# En las listas i_dentro i_fuera guardo los indices de los puntos que quedan dentro y fuera del semicirculo
i_dentro = [] 
i_fuera = []

for i in range(1000):
    x[i] = 2.0*(random.random()-0.5) # Numero aleatorio entre -1 y 1
    y[i] = random.random() # Numero aleatorio entre 0 y 1
    
    r[i] = (x[i]**2 + y[i]**2)**0.5 # Calcular el radio que representa el par x,y
    if r[i] <=1.0: # Si el radio es menor a 1, el punto esta dentro del semicirculo
        i_dentro.append(i)
    else: # si no, esta fuera del semicirculo
        i_fuera.append(i)

# Completar la grafica        
title('Semi-circulo unitario frecuentista')
# Graficar puntos dentro y fuera con diferente color
scatter(x[i_fuera],y[i_fuera],color='red', label='encima') 
scatter(x[i_dentro],y[i_dentro],color='blue', label='debajo')
axis('equal')
xlim([-1.1,1.1]) # Mostrar solo el dominio donde cree puntos
xlabel('x'); ylabel('y')
legend(loc=2,framealpha=0.6)
```

```python
#############################
#                           #
#          4                #
#                           #
#############################

def distribucion_dos_campanas(x):
    # amplitud y parte exponencial de la primera campana en la ecuacion
    a1 = 0.3/sqrt(0.0025*pi)
    e1 = exp(-((x-5.0)/0.05)**2)
    
    # amplitud y parte exponencial de la segunda campana en la ecuacion
    a2 = 0.7/sqrt(pi)
    e2 = exp(-((x-2.0)/1.0)**2)
    
    return a1*e1 + a2*e2
    

x = arange(0.0,6.0,0.05) # Crear puntos desde 0 hasta 6, cada 0.05 para hacer la grafica de la distribucion
y = distribucion_dos_campanas(x) # Calcular los valores y de la distribucion


figure()
plot(x,y,c='k',lw=2,label='dist2') # Graficar la linea continua de la distribucion


N = 1000 # Cantidad de datos
# Crear arreglos de x,y con N posiciones para guardar las muestras aleatorias
x = zeros(N) 
y = zeros(N)
# En las listas i_dentro i_fuera guardo los indices de los puntos que quedan dentro y fuera de la distribucion
i_dentro = []
i_fuera = []

for i in range(N):
    x[i] = 6.0*(random.random()) # Numero aleatorio entre 0 y 6
    y[i] = 3.5*random.random() # Numero aleatorio entre 0 y 3.5
    
    if y[i] <= distribucion_dos_campanas(x[i]) : 
        i_dentro.append(i) # Si el y aleatorio es menor que el y que correspone al x aleatorio, esta dentro de la distribucion
    else: # si no, esta afuera
        i_fuera.append(i)

# Completar la grafica        
title('Distribucion problema para frecuentista')
# Graficar puntos adentro y afuera de la distribucion como en el enunciado
scatter(x[i_fuera],y[i_fuera],color='red', label='encima')
scatter(x[i_dentro],y[i_dentro],color='blue', label='debajo')
axis('equal') # Especficiar relacion de easpecto 1:1
xlabel('x'); ylabel('y')
legend(loc=2,framealpha=0.6)
```

```python
#############################
#                           #
#          6                #
#                           #
#############################

def distribucion_dos_campanas(x):
    # amplitud y parte exponencial de la primera campana en la ecuacion
    a1 = 0.3/sqrt(0.0025*pi)
    e1 = exp(-((x-5.0)/0.05)**2)
    
    # amplitud y parte exponencial de la segunda campana en la ecuacion
    a2 = 0.7/sqrt(pi)
    e2 = exp(-((x-2.0)/1.0)**2)
    
    return a1*e1 + a2*e2

def likelihood(x_old, x_new): # Funcion de likelihood que se usa en Metropolis-Hastings
    # Para el x_viejo y el x_nuevo calculo la distribucion de probabilidad
    # Divido la nueva sobre la vieja
    # Como es distribucion de probabilidad, entre mayor sea el numero "y" significa que
    # el punto tiene mayor probabilidad en la distribucion.
    # Entonces, me interesa que cada vez los puntos tengan mayor probabilidad.
    # En esta funcion de likelihood, si es mayor que 1 significa que mejoro la probabilidad
    # entonces debo tomar el punto. Si no, significa que la probabilidad no mejoro y 
    # necesito aplicar otro procedimiento para decidir si acepto el nuevo punto o no
    y_old = distribucion_dos_campanas(x_old)
    y_new = distribucion_dos_campanas(x_new)
    return y_new/y_old
    
    
x = arange(0.0,6.0,0.05) # Crear puntos desde 0 hasta 6, cada 0.05 para hacer la grafica de la distribucion
y = distribucion_dos_campanas(x) # Calcular los valores y de la distribucion

figure()
plot(x,y,c='k',lw=2,label='dist2') # Graficar la linea continua de la distribucion


N = 100000 # Numero de puntos
x = [] # Lista para guardar los valores de x muestreados con Metropolis-Hastings
x.append( (random.normal(0.0,1.0)+3) ) # El primer numero que guardo es un numero aleatorio
# muestreado con una gausiana centrada en la parte media del dominio (3) y con desviacion estandar
# de 1 para que el 96% de los posibles numeros este entre 1 y 5

for i in range(N-1):
    xn = random.normal(x[i],0.4) # nuevo numero aleatorio muestreado con gausiana centrada en el punto
    # que ya tenia y con desviacion estandar pequeña (en comparacion al dominio) para que los pasos no sean muy grandes
    likelhd = likelihood(x[i],xn) # Calcular likelihood con el punto viejo y el punto nuevo
    
    if likelhd >= 1.0:
        # Como se explico en el likelihood, si el valor es mayor a 1 debo aceptar el nuevo numero
        x.append(xn)
    else: # si no, lanzo un nuevo numero aleatorio entre 0 y 1
        # Voy a aceptar el nuevo valor con una probabilidad del valor del likelihood (ya se que es un numero entre 0 y 1)
        # Para esto, lanzo un nuevo numero aleatorio entre 0 y 1 con distribucion uniforme
        # Si el nuevo numero aleatorio beta es menor que el likelihood, acepto el nuevo valor
        # si no, me quedo con el valor anterior
        beta = np.random.random()
        if likelhd > beta:
            x.append(xn)
        elif likelhd < beta:
            x.append(x[i])

# Como x era una lista lo convierto a arreglo
# x = array(x)            
# Retiro los primeros 500 datos
x = x[500:]

# Completar grafica
title('Histograma frecuentista')
hist(x, bins=80, normed=1,color='orange',edgecolor='orange',label='Muestro Metropolis-Hastings') # Poner 80 bins (cajones) para que se vea mejor
axis('equal') # ASegurar relacion de aspecto 1:1
xlabel('x'); ylabel('y')
legend(loc=2,framealpha=0.6)
xlim([0,6])
```
